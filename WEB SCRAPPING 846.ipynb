```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests as re
```

```python
url="https://dir.indiamart.com/search.mp?ss=hp+laptop&src=as-popular%7Ckwd%3Dlaptop%7Cpos%3D4%7Ccat%3D-2%7Cmcat%3D-2%7Ckwd_len%3D6%7Ckwd_cnt%3D1&v=4&mcatid=&catid=&tags=stype:attr=1-br|qr_nm:gd|res:RC5|com-cf:nl|ptrs:na|ktp:N0|mc:181436|mtp:Brn|qry_typ:P|lang:en|wc:2|flavl:10|cs:10164"
```

```python
response=re.get(url)
```

```python
response
```

```python
bc=bs(response.text)
```

```python
laptop_name=bc.select(".elps2 .cardlinks")
```

```python
laptop_name
```

```python
laptop_data=pd.DataFrame(laptop_name,columns=["name"])
```

```python
laptop_data
```

```python
laptop_price=bc.select(".price")
```

```python
laptop_data["price"]=laptop_price
```

```python
laptop_data
```

```python
laptop_address=bc.select(".elps1")
```

```python
laptop_data["Address"]=laptop_address
```
